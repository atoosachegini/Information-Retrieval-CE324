 بنابراین استفاده‌ی مستقیم از آنها معمولا نابهینه بوده و امکان کاهش ابعادی آنها بدون از دست رفتن زیاد اطلاعات وجود دارد
به همین منظور یک pipeline جداگانه برای محاسبه‌ی بردار‌های TF-IDF و کاهش ابعاد آنها به کمک 
<a href="https://en.wikipedia.org/wiki/Singular_value_decomposition">Singular Value Decomposition (SVD)</a>

در اینجا به کمک pipeline 
پیاده‌سازی شده، دادگان پروژه را برای بخش‌های بعدی آماده‌می‌کنیم. در صورت نیاز می‌توانید پیاده‌سازی pipeline و یا پارامتر‌های آن‌را برای بهبود کیفیت بردار‌های یافته شده تغییر دهید.

```python
# setup preprocessor
preprocessor.set_params(
    vectorizer__n_components=256, # dimensionality of the word vectors
    vectorizer__w2v={
        'iter':128, # number of training iterations of the w2v model
        'min_count':2, # all words with total frequency lower thaک.n this will be ignored
    },
)
preprocessor.fit(train_data)
train_data = preprocessor.transform(train_data)
test_data = preprocessor.transform(test_data)

train_data.head(2) # see how the preprocessed data looks!
```

   هم برای اعتبار‌سنجی عمل‌کرد الگوریتم‌های یادگیری ماشین و انتخاب بهینه‌تر پارامتر‌های آنها وجود دارد و به جای این‌کار می‌توانید از آنها استفاده‌کنید.

```python
for i in range(len(train_data)):
    train_data['vec_1'][i] = ast.literal_eval(train_data['vec_1'][i])
    temp = train_data['vec_2'][i][1:-1].split()
    for j in range(len(temp)):
        temp[j] = float(temp[j])
    train_data['vec_2'][i] = temp
for i in range(len(test_data)):
    test_data['vec_1'][i] = ast.literal_eval(test_data['vec_1'][i])
    temp1 = test_data['vec_2'][i][1:-1].split()
    for j in range(len(temp1)):
        temp1[j] = float(temp1[j])
    test_data['vec_2'][i] = temp1
```

     برای یادآوری شیوه‌ی عملکرد این الگوریتم می‌توانید

را بررسی کنید.

    </font>

```python
Kmeansmodel1 = clustering.KMeans(cluster_count=18, max_iteration=40000, tolerance=1, distance_type='cos')
Kmeansmodel1.fit(train_data['vec_1'])
Kmeanspredicted1 = Kmeansmodel1.predict(train_data['vec_1'])

Kmeansmodel2 = clustering.KMeans(cluster_count=11, max_iteration=1000, tolerance=1, distance_type='cos')
Kmeansmodel2.fit(train_data['vec_2'])
Kmeanspredicted2 = Kmeansmodel2.predict(train_data['vec_2'])

Kmeanspurity1 = clustering.purity(train_data['genre'], Kmeanspredicted1)
Kmeanspurity2 = clustering.purity(train_data['genre'], Kmeanspredicted2)
print('purity of Kmeans with vec_1 : ' + str(Kmeanspurity1))
print('purity of Kmeans with vec_2 : ' + str(Kmeanspurity2))
```

    <h3>Gaussian Mixture Models
        (۸ نمره)
    </h3>
    الگوریتم GMM را در 
    <code style="font-size:0.9em">src.clustering.models</code>
    با استقاده از کتابخانه‌های موجود (مانند

    پیاده‌سازی کرده و مقدار بهینه برای هایپرپرامتر‌های این خوشه‌بند (تعداد خوشه‌ها، تعداد گام‌ها و ...) را بدست‌ آورده و در نهایت بهترین تنظیمات و میزان عملکرد مربوط به آن را در کد زیر گزارش کنید. در صورتی که به شیوه‌ی پیاده‌سازی این الگوریتم علاقه دارید می‌توانید به این <a href="https://towardsdatascience.com/gaussian-mixture-modelling-gmm-833c88587c7f">مقاله</a> مراجعه کنید.
    </font>
</div>

```python
GMMmodel1 = clustering.GMM(cluster_count=11, max_iteration=10000, covariance_type='full')
GMMmodel1.fit(train_data['vec_1'])
GMMpredicted1 = GMMmodel1.predict(train_data['vec_1'])

GMMmodel2 = clustering.GMM(cluster_count=11, max_iteration=10000, covariance_type='spherical')
GMMmodel2.fit(train_data['vec_2'])
GMMpredicted2 = GMMmodel2.predict(train_data['vec_2'])

GMMpurity1 = clustering.purity(train_data['genre'], GMMpredicted1)
GMMpurity2 = clustering.purity(train_data['genre'], GMMpredicted2)
print('purity of GMM with vec_1 : ' + str(GMMpurity1))
print('purity of GMM with vec_2 : ' + str(GMMpurity2))
```

    <h2>مصور سازی (۱۰ نمره امتیازی)

    در این قسمت، مشابه قسمت قبل، از شما خواسته شده‌است تا خوشه‌بندی‌های بدست آمده در قسمت قبل را در دو بعد مصورسازی کرده تا بتوانید درک بهتری از شیوه‌ی عملکرد این خوشه‌بندها داشته باشید. سعی کنید عملکرد خوشه‌بند‌ها را با یکدیگر مقایسه کرده و نتایج خود را گزارش کنید. برای رسم نمودارها ابتدا بایستی با استفاده از روش‌های کاهش ابعاد

    مجموعه داده‌ها را به فضای برداری دو یا سه بعدی ببرید تا امکان مصورسازی فراهم شود. همچنین استفاده از کتابخانه‌های موجود برای کاهش ابعاد و مصورسازی مانعی ندارد.
    </font>
</div>

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
train_vec1 = []
for i in range(len(train_data['vec_1'])):
    train_vec1.append(train_data['vec_1'][i])
train_vec1 = np.array(train_vec1)
train_vec2 = []
for i in range(len(train_data['vec_2'])):
    train_vec2.append(train_data['vec_2'][i])
train_vec2 = np.array(train_vec2)
train_vec1_2d = tsne.fit_transform(train_vec1)
train_vec2_2d = tsne.fit_transform(train_vec2)
```

```python
targets = [0, 1, 2, 3, 4, 5, 6, 7]
color = 'r', 'g', 'b', 'c', 'm', 'y', 'k', 'orange'
target_ids = range(len(targets))
#Kmeans
print('Kmeans for vec_1')
plt.figure(figsize=(6, 5))
for i, c, label in zip(target_ids, color, targets):
    print(i, c, label)
    plt.scatter(train_vec1_2d[np.array(Kmeanspredicted1) == i, 0], train_vec1_2d[np.array(Kmeanspredicted1) == i, 1], c=c, label=label)
plt.legend()
plt.show()
print('Kmeans for vec_2')
plt.figure(figsize=(6, 5))
for i, c, label in zip(target_ids, color, targets):
    plt.scatter(train_vec2_2d[np.array(Kmeanspredicted2) == i, 0], train_vec2_2d[np.array(Kmeanspredicted2) == i, 1], c=c, label=label)
plt.legend()
plt.show()

# #KNN
print('GMM for vec_1')
plt.figure(figsize=(6, 5))
for i, c, label in zip(target_ids, color, targets):
    plt.scatter(train_vec1_2d[GMMpredicted1 == i, 0], train_vec1_2d[GMMpredicted1 == i, 1], c=c, label=label)
plt.legend()
plt.show()
print('GMM for vec_2')
plt.figure(figsize=(6, 5))
for i, c, label in zip(target_ids, color, targets):
    plt.scatter(train_vec2_2d[GMMpredicted2 == i, 0], train_vec2_2d[GMMpredicted2 == i, 1], c=c, label=label)
plt.legend()
plt.show()

#Hierarchical
print('Hierarchical for vec_1')
plt.figure(figsize=(6, 5))
for i, c, label in zip(target_ids, color, targets):
    plt.scatter(train_vec1_2d[Hpredicted1 == i, 0], train_vec1_2d[Hpredicted1 == i, 1], c=c, label=label)
plt.legend()
plt.show()
print('Hierarchical for vec_2')
plt.figure(figsize=(6, 5))
for i, c, label in zip(target_ids, color, targets):
    plt.scatter(train_vec2_2d[Hpredicted2 == i, 0], train_vec2_2d[Hpredicted2 == i, 1], c=c, label=label)
plt.legend()
plt.show()
```
